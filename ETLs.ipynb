```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(df_hechos_clean_1):
    # Drop column: 'comuna'
    df_hechos_clean_1 = df_hechos_clean_1.drop(columns=['comuna'])
    # Convert text to lowercase in columns: 'id_hecho', 'longitud' and 3 other columns
    df_hechos_clean_1['id_hecho'] = df_hechos_clean_1['id_hecho'].str.lower()
    df_hechos_clean_1['longitud'] = df_hechos_clean_1['longitud'].str.lower()
    df_hechos_clean_1['latitud'] = df_hechos_clean_1['latitud'].str.lower()
    df_hechos_clean_1['victima'] = df_hechos_clean_1['victima'].str.lower()
    df_hechos_clean_1['acusado'] = df_hechos_clean_1['acusado'].str.lower()
    return df_hechos_clean_1

df_hechos_clean_2 = clean_data(df_hechos_clean_1.copy())
df_hechos_clean_2.head()
```

```python
from datetime import datetime

# Definir la función para convertir a formato de tiempo
def convertir_a_time(x):
    if isinstance(x, str):
        try:
            return datetime.strptime(x, "%H:%M:%S").time()
        except ValueError:
            return None
    elif isinstance(x, datetime):
        return x.time()
    return x

# Aplicar la función a la columna 'hora' del DataFrame df_hechos
df_hechos_clean_2['hora'] = df_hechos_clean_2['hora'].apply(convertir_a_time)

# Verificar el resultado
print(df_hechos_clean_2['hora'].head())
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(df_victimas):
    # Drop rows with missing data in columns: 'sexo_victima', 'victima', 'edad_victima'
    df_victimas = df_victimas.dropna(subset=['sexo_victima', 'victima', 'edad_victima'])
    # Drop columns: 'tipo_de_dato', 'fecha_fallemiento' and 2 other columns
    df_victimas = df_victimas.drop(columns=['tipo_de_dato', 'fecha_fallemiento', 'rol_victima', 'gravedad'])
    # Convert text to lowercase in columns: 'id_hecho', 'victima', 'sexo_victima'
    df_victimas['id_hecho'] = df_victimas['id_hecho'].str.lower()
    df_victimas['victima'] = df_victimas['victima'].str.lower()
    df_victimas['sexo_victima'] = df_victimas['sexo_victima'].str.lower()
    # Remove leading and trailing whitespace in columns: 'id_hecho', 'victima', 'sexo_victima'
    df_victimas['id_hecho'] = df_victimas['id_hecho'].str.strip()
    df_victimas['victima'] = df_victimas['victima'].str.strip()
    df_victimas['sexo_victima'] = df_victimas['sexo_victima'].str.strip()
    # Rename column 'aaaa' to 'año'
    df_victimas = df_victimas.rename(columns={'aaaa': 'año'})
    return df_victimas

df_victimas_clean = clean_data(df_victimas.copy())
df_victimas_clean.head()
```

```python
# Guardar el DataFrame en Excel
df_victimas_clean.to_excel('victimas.xlsx', sheet_name='victimas', index=False)

print("Datos guardados exitosamente en 'victimas.xlsx'")
```
